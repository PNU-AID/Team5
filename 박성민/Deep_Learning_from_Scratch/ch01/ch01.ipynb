## practice class

```python
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
    
    def hello(self):
        print("HEllo " + self.name + "!")
    
    def goodbye(self):
        print("Good-bye " + self.name + "!")
    
m = Man("David")
m.hello()
m.goodbye()
```

```python
y = np.array([2.0, 4.0, 6.0])
# element-wise add
print(x + y)
# element-wise sub
print(x - y)
# element-wise product
print(x * y)
# element-wise div
print(x / y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# 데이터 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)

#그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") # cos 함수는 점선으로 그리기
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title('sin & cos') # 제목
plt.legend()
plt.show()
```
